```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Data/Reason/general reason for moving.csv.gz', compression='gzip')

df = df.rename(columns={col: col.replace('\n', '').replace('.', '') for col in df.columns}).fillna(0)

df = df.loc[:6, :]

df
```

```python
# create line graph in paper
graph_data = df[['Movers in the United States', 'Total'] + [col for col in df if '-related' in col]]
graph_data = graph_data.rename(columns={col: col.replace('-related reason', '').strip('2') for col in graph_data.columns} | {'Movers in the United States': 'Year'})
graph_data = graph_data/1000
graph_data['Year'] = graph_data['Year'] * 1000
graph_data.to_csv('./Output/Reason/moving_graph.csv', index=False)
```

```python
table_data = df[[col for col in df.columns if col not in graph_data.columns]].set_index('Movers in the United States').T.replace('-', 0).astype(float)
table_data
```

```python
family_reason = table_data.index[1:4]
employment_reason = table_data.index[5:10]
housing_reason = table_data.index[11:17]
other_reason = table_data.index[18:24]

housing_reason
```

```python
# format and calculate percentage
idx_order = []
new_idx = []

idx_order.append('Family-related reasons')
new_idx.append('Family-related reasons')
for reason in family_reason:
    idx_order.append(reason)
    idx_order.append(reason+' percentage')
    new_idx.append(reason)
    new_idx.append('%')
    table_data.loc[reason + ' percentage'] = table_data.loc[reason]/table_data.loc['Family-related reasons']*100

idx_order.append('Employment-related reasons')
new_idx.append('Employment-related reasons')
for reason in employment_reason:
    idx_order.append(reason)
    idx_order.append(reason+' percentage')
    new_idx.append(reason)
    new_idx.append('')
    table_data.loc[reason + ' percentage'] = table_data.loc[reason]/table_data.loc['Employment-related reasons']*100

idx_order.append('Housing-related reasons')
new_idx.append('Housing-related reasons')
for reason in housing_reason:
    idx_order.append(reason)
    idx_order.append(reason+' percentage')
    new_idx.append(reason)
    new_idx.append('')
    table_data.loc[reason + ' percentage'] = table_data.loc[reason]/table_data.loc['Housing-related reasons']*100

idx_order.append('Other-related reasons')
new_idx.append('Other-related reasons')
for reason in other_reason:
    idx_order.append(reason)
    idx_order.append(reason+' percentage')
    new_idx.append(reason)
    new_idx.append('')
    table_data.loc[reason + ' percentage'] = table_data.loc[reason]/table_data.loc['Other-related reasons']*100

table_data = table_data.loc[idx_order]
table_data
```

```python
tmp = table_data.copy()
tmp
```

```python
#create latex file
tmp = table_data.copy()

for idx in table_data.index:
    if 'percentage' in idx:
        tmp.loc[idx] = np.round(table_data.loc[idx], 2)

tmp = tmp.astype(str)

for idx in table_data.index:
    if 'percentage' in idx:
        tmp.loc[idx] += '%'
    else:
        tmp.loc[idx] = tmp.loc[idx].str.slice(stop=-2)
        
tmp[range(2016, 2023)].to_latex('./Output/Reason/moving_composition.tex')
```
